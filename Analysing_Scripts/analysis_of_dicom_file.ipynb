```python
import os
import pydicom
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pydicom import dcmread
# Define the path to the data folder
root = os.path.dirname(os.getcwd())
dicom_file= os.path.join(root, 'data', 'IS20250115_171841_9465_61003253.dcm')
x= dcmread(dicom_file)
print(x)
```

```python
import matplotlib.pyplot as plt
pixel_data = x.pixel_array
plt.imshow(pixel_data, cmap='gray')
plt.title('DICOM Image')
plt.show()
```
